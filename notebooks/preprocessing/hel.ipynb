# **HEL Dataset Preprocessing**

```python
from src import utilities
from src import preprocessing
from src import plotting

hel_ds_path = '../../datasets/CreditRiskAnalytics/hel.csv'
hel_prep_ds_path = '../../datasets/preprocessed/hel/'
```

```python
utilities.create_directory(hel_prep_ds_path)
```

```python
data = utilities.read_csv(hel_ds_path, dtype={'BAD':'bool', 'REASON':'category',
                                                    'JOB':'category', 'LOAN':'float64'})
data
```

```python
data.info()
```

#### Plot features data types.

```python
plotting.plot_dtypes_hist(data, save_path=hel_prep_ds_path)
```

#### Data cleaning.

```python
before_data_cleaning_count = len(data)
```

#### Rename features labels to have more meaningful names.

```python
data.rename(columns={'BAD':'defaulted',
                     'LOAN':'loan-amount',
                     'MORTDUE':'existing-amount-due',
                     'VALUE':'curr-prop-val',
                     'REASON':'loan-reason',
                     'JOB':'job',
                     'YOJ':'years-at-job',
                     'DEROG':'derogatories',
                     'DELINQ':'delinquencies',
                     'CLAGE':'oldest-credit-line',
                     'NINQ':'credit-inquiries',
                     'CLNO':'credit-lines',
                     'DEBTINC':'debt-to-income'}, inplace=True)
```

#### Plot class distribution.

```python
plotting.plot_class_distr_hist(data, save_path=hel_prep_ds_path)
```

#### Plot missing values histogram.

```python
plotting.plot_missing_hist(data, figsize=(8,8), save_path=hel_prep_ds_path)
```

#### Apply optimal binninf and WoE.

```python
data, features_scores = preprocessing.optbinning_woe(data=data, solver="cp",
                                                     outlier_detector="range",
                                                     save_path=hel_prep_ds_path,
                                                     verbose=True)
```

```python
data
```

```python
plotting.plot_numerical_boxplots(data, size=(15, 6), save_path=hel_prep_ds_path)
```

```python
plotting.plot_numerical_hist_kde(data, size=(15,5), save_path=hel_prep_ds_path)
```

#### Plot samples count before and after Data cleaning.

```python
after_data_cleaning_count = len(data)
labels = ['Before Data Cleaning', 'After Data Cleaning']
values = [before_data_cleaning_count, after_data_cleaning_count]
plotting.plot_hist(labels, values, '', '', 'Samples', figsize=(8,5), rotated_ticks=False,
                   save_path=hel_prep_ds_path + 'before_after.pdf')
```

#### Features IV scores

```python
plotting.plot_features_scores(features_scores, '')
```

#### Features correlation

```python
f_corr = preprocessing.features_correlation(data=data)
plotting.plot_heatmap(f_corr, figsize=(12, 12), save_path=hel_prep_ds_path, dpi=100)
```

#### Drop redundant features using IV score.

```python
to_be_dropped = ['delinquencies']
data.drop(to_be_dropped, axis=1, inplace=True)
for key in to_be_dropped:
    del features_scores[key]
```
